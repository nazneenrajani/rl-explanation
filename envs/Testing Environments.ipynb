```python
def make_env(seeds, rank=0):
    env_name = 'MiniGrid-BallFindingTrain-v0'
    env = gym.make(env_name)
    env = FullyObsWrapper(env)
#     env = ImageChannelSwapWrapper(env)
#     env = MiniGridRewardWrapper(env)
    env = MiniGridTimeLimitWrapper(env, 5000)
#     print (rank)
    env = ReseedWrapper(env, seeds, rank % len(seeds))
    env = PredicateWrapper(env, [])
    return env
    
env = make_env([1])
print (env.reset().shape)
im = env.render(mode='rgb_array')
plt.imshow(im)
print (env.observation_space)

from stable_baselines.common.vec_env import DummyVecEnv
# from stable_baselines.deepq.policies import MlpPolicy
from stable_baselines.common.policies import MlpPolicy
from stable_baselines import DQN, PPO2
from stable_baselines.common.vec_env import SubprocVecEnv

env = SubprocVecEnv([lambda: env for _ in range(16)])

model = PPO2(MlpPolicy, env, verbose=1)
print ("Learning")
model.learn(total_timesteps=2000000)
```

```python
model.save("ppo2__")
```

```python
# Enjoy trained agent
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    env.render()
```
